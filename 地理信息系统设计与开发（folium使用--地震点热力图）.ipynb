```python
m = folium.Map(
    #location=[lat,lng],
    location=[23.529162,21.354245],
    zoom_start=1,
    #crs='EPSG4326',
    #tiles='Stamen Terrain', #更改了展开图样式
    #tiles="OpenStreetMap",
    #tiles='Stamen Toner',
    #tiles='http://wprd01.is.autonavi.com/appmaptile?x={x}&y={y}&z={z}&lang=zh_cn&size=1&scl=1&style=7',
    tiles='http://webst02.is.autonavi.com/appmaptile?style=6&x={x}&y={y}&z={z}',
    #tiles='http://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=7&x={x}&y={y}&z={z}',
    #tiles='http://www.google.cn/maps/vt?lyrs=s@820&gl=cn&x={x}&y={y}&z={z}&zmax=18&zmin=0',
    #tiles='http://www.google.cn/maps/vt?lyrs=s@189&gl=cn&x={x}&y={y}&z={z}',
    #tiles="http://www.google.cn/maps/vt?lyrs=m@189&gl=cn&x={x}&y={y}&z={z}",
    #tiles="http://thematic.geoq.cn/arcgis/rest/services/ThematicMaps/administrative_division_boundaryandlabel/MapServer/tile/{z}/{y}/{x}",
    #tiles= "http://cache1.arcgisonline.cn/arcgis/rest/services/ChinaOnlineCommunity/MapServer/tile/{z}/{y}/{x}",
    attr='default'
)
m
```

```python
ds = ogr.Open("test3.shp")
lyr=ds.GetLayer()
result = ds.ExecuteSQL("select * from test3 where longitude > 66 AND longitude < 76 AND latitude > 34 AND latitude < 44")
print("result:",result.GetFeatureCount())
lyr.SetAttributeFilter("magnitude > 5")
print(lyr.GetFeatureCount())
```

```python
data1=[]
for fea in lyr:
    record=[]
    mag=fea.GetField('magnitude')

    geom=fea.geometry()
    #print("geom=",geom)
    lon=geom.GetX()

    lat=geom.GetY()

    record.append(lat)
    record.append(lon)
    record.append(mag)
    data1.append(record)
    folium.Marker(
        location=[lat, lon],
        #popup='Mt. Hood Meadows',
        #icon=folium.Icon(icon='cloud')
    ).add_to(m)

m
```

```python
print(data1)
```

```python
HeatMap(data1).add_to(m) 
m
```

```python
file_path = "earthquakeHeatMap.html"
# 保存为html文件
m.save(file_path)
# 默认浏览器打开
webbrowser.open(file_path)
```
