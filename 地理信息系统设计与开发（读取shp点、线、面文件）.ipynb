# 地理信息系统设计与开发（读取shp点、线、面文件）

```python
import ogr
import osr
import json
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
driver = ogr.GetDriverByName('ESRI Shapefile')
```

## 面

```python
ds=ogr.Open(r'D:\云南省1比100万\云南省地州市.shp')
print("ds.type:",ds)
```

```python
layer_polygon=ds.GetLayer(0)
print("layer:",layer_polygon)
```

```python
for row in layer_polygon:
	print(row.GetField(2))
	geom=row.geometry()
	ring=geom.GetGeometryRef(0)
	coords=ring.GetPoints()
	x,y=zip(*coords)
	plt.plot(x,y,'k')
	plt.fill(x,y,'p')
```

```python
print("geometry:",geom)
```

```python
print("ring:",ring)
```

```python
print("geometryname:",geom.GetGeometryName())
```

```python
print("coords:",coords)
```

## 线

```python
ds=ogr.Open(r'D:\云南省1比100万\云南铁路.shp')
layer_line=ds.GetLayer(0)
```

```python
for feature in layer_line:
	geom=feature.geometry()
	coords=geom.GetPoints()
	if coords is None:
		continue
	x,y=zip(*coords)
	plt.plot(x,y,'r')
	#plt.text(x[0], y[0], 'test')
print("geometry:",geom)
```

```python
print("geometryname:",geom.GetGeometryName())
```

## 点

```python
ds=ogr.Open(r'D:\云南省1比100万\云南县城点.shp')
layer=ds.GetLayer(0)
for feature in layer:
	geom=feature.geometry()
	x=geom.GetX()
	y=geom.GetY()
	plt.plot(x,y,'o',markersize=5,color='k')
	#plt.text(x,y,"test")
```

```python
print("geometryname:",geom.GetGeometryName())
```
